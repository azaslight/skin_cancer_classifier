```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Путь к изображению после загрузки
img_path = 'ISIC_0024306.jpg'  # Имя файла, которое вы загрузили

# Загрузка изображения
img = Image.open(img_path)

# Показываем изображение "До обработки"
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.title("До обработки")
plt.axis('off')  # Убираем оси
plt.show()

# Применяем обработку: изменение размера, нормализация
img = img.resize((64, 64))  # Изменение размера
img = img.convert('RGB')  # Преобразование в RGB
img_array = np.array(img) / 255.0  # Нормализация

# Преобразование обратно в изображение для отображения
img_processed = Image.fromarray((img_array * 255).astype(np.uint8))  # Преобразование обратно в изображение

# Показываем обработанное изображение "После обработки"
plt.figure(figsize=(6,6))
plt.imshow(img_processed)
plt.title("После обработки")
plt.axis('off')  # Убираем оси
plt.show()
```
